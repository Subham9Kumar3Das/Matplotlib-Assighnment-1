### Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.

Matplotlib is a popular Python library used for creating various types of plots and graphs. It provides a flexible and user-friendly way to visualize data and is widely used in data analysis, scientific research, and data visualization tasks.

Here are five types of plots that can be created using the Pyplot module of Matplotlib:

* Line Plot: Line plots are used to visualize data points as a series of connected line segments. They are useful for showing trends and changes in data over time.

* Bar Plot: Bar plots represent data using rectangular bars, where the height or length of each bar corresponds to the value of the data it represents. They are commonly used to compare and display categorical data.

* Histogram: Histograms are used to visualize the distribution of a continuous dataset. They display data as bars grouped into intervals (bins), showing the frequency or count of data points falling into each bin.

* Scatter Plot: Scatter plots are used to visualize the relationship between two variables. They display individual data points as dots on a two-dimensional plane, making it easy to identify patterns or correlations.

* Pie Chart: Pie charts represent data as a circle divided into slices, where each slice's size represents the proportion of a whole. They are effective for showing the composition of a dataset, such as the percentage distribution of different categories within a whole.

### Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
#Note: Also add title, xlabel, and ylabel to the plot.
```

```python
import matplotlib.pyplot as plt 
```

```python
plt.title('Scatter plot')
plt.xlabel('x-axis--->')
plt.ylabel('y-axis--->')
plt.scatter(x,y,edgecolors='r')
plt.show()
```

### Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

### Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

Ans:-A box plot, also known as a box-and-whisker plot, is a graphical representation of a dataset's distribution. It shows important statistics about the data, including the median, quartiles, and potential outliers. Box plots are used to visually summarize and compare the spread, central tendency, and skewness of data.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
plt.figure(figsize=(8,4))
plt.boxplot([box1,box2],labels=['box1','box2'])
plt.title('Box plot')
plt.xlabel('x-axis-->')
plt.ylabel('y-axis-->')
plt.show()
```
